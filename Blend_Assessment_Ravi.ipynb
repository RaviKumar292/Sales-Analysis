**`Feature Engineering`**

````python
df1 = sales_merged.copy()
df1.columns = (
    df1.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace(".", "_")
)
````

````python
sku_cols = [c for c in df1.columns if 'sku' in c.lower()]
df1['sku_final'] = df1[sku_cols].bfill(axis=1).iloc[:,0]
df1 = df1.drop(columns=sku_cols)
df1 = df1.rename(columns={'sku_final': 'sku'})
````

````python
cat_cols = [c for c in df1.columns if 'category' in c.lower() and c != 'category']
if cat_cols:
    df1['category'] = df1[cat_cols].bfill(axis=1).iloc[:,0]
    df1 = df1.drop(columns=cat_cols, errors='ignore')
else:
    df1['category'] = df1.get('catalog', 'Unknown')

size_cols = [c for c in df1.columns if 'size' in c.lower()]
df1['size'] = df1[size_cols].bfill(axis=1).iloc[:,0]
df1 = df1.drop(columns=size_cols)

drop_cols = [c for c in df1.columns if 'index' in c or 'unnamed' in c]
df2 = df1.drop(columns=drop_cols, errors='ignore')
````

````python
df3 = df2.copy()

# Converting date column to datetime
df3['order_date'] = pd.to_datetime(df3['date'], format='%m-%d-%y', errors='coerce')

# Droping invalid dates
df3 = df3[df3['order_date'].notna()]
df3 = df3.drop(columns=['date'])

# Derived time features
df3['year'] = df3['order_date'].dt.year
df3['month'] = df3['order_date'].dt.month
df3['day'] = df3['order_date'].dt.day
df3['quarter'] = df3['order_date'].dt.quarter
df3['week_num'] = df3['order_date'].dt.isocalendar().week.astype(int)
df3['day_of_week'] = df3['order_date'].dt.day_name()

df3.head()
````

````python
df3['qty'] = pd.to_numeric(df3['qty'], errors='coerce').astype('Int64')
df3['amount'] = (
    df3['amount']
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df3['amount'] = pd.to_numeric(df3['amount'], errors='coerce')
````

````python
# Identify all numeric-like columns
num_cols = [
    'qty', 'amount', 'rate', 'tp', 'mrp_old', 'final_mrp_old',
    'ajio_mrp', 'amazon_mrp', 'amazon_fba_mrp', 'flipkart_mrp',
    'limeroad_mrp', 'myntra_mrp', 'paytm_mrp', 'snapdeal_mrp',
    'tp_1', 'tp_2', 'stock'
]

for col in num_cols:
    if col in df3.columns:
        df3[col] = (
            df3[col]
            .astype(str)
            .str.replace("₹", "", regex=False)
            .str.replace(",", "", regex=False)
            .str.replace(" ", "", regex=False)
            .str.strip()
        )
        df3[col] = pd.to_numeric(df3[col], errors='coerce')
````

````python
if 'qty' in df3.columns:
    df3['qty'] = df3['qty'].astype('Int64')

if 'stock' in df3.columns:
    df3['stock'] = df3['stock'].astype('Int64')
````

**`Natural Language to SQL Pipeline`**

````python
def validate_sql(sql):
    sql_lower = sql.lower()

    if "delete" in sql_lower or "update" in sql_lower or "insert" in sql_lower:
        return False
    if "drop" in sql_lower or "alter" in sql_lower:
        return False

    # Must start with SELECT
    if not sql_lower.strip().startswith("select"):
        return False

    return True
````

````python
def nl_to_sql(question):
    prompt = f"""
    Convert the question into valid DuckDB SQL.

    Use table: orders

    Available columns:
    {", ".join(df3.columns)}

    Rules:
    If filtering by year, use the 'year' column
    Use SUM(qty) when referring to selling the most
    Exclude NULL or missing sku values
    Order numerically, not alphabetically
    Return ONLY the SQL

    Question: "{question}"
    """
    response = model.generate_content(prompt)
    return response.text.strip().replace("```sql", "").replace("```", "")
````

````python
def run_sql(sql):
    try:
        df4 = con.execute(sql).fetchdf()
        return df4
    except Exception as e:
        return f"SQL Error: {e}"
````

````python
def table_to_insight(question, df4):
    prompt = f"""
    You are a business insights analyst.
    Use the table below to answer the question:

    QUESTION: {question}
    DATA:
    {df4.to_json()}

    Provide a short 3–5 sentence answer.
    Include numbers in your explanation.
    """
    return model.generate_content(prompt).text
````

````python
def ask_insight(question):
    sql = nl_to_sql(question)
    # sql = fix_date_functions(sql)
    print("Generated SQL:\n", sql)

    df5 = run_sql(sql)
    if df5.empty:
      return "No meaningful data available to answer the question."
    print("\nQuery Result:\n", df5)

    if isinstance(df5, str):
        return df5

    answer = table_to_insight(question, df5)
    return answer
````

**`Enhanced NL and SQL with Usability`**

````python
def execute_sql(sql):
    try:
        df = con.execute(sql).fetchdf()
        return df
    except Exception as e:
        return f"SQL Error: {str(e)}"
````

````python
def ask_sales_question(question):
    sql = nl_to_sql(question)
    print("Generated SQL:\n", sql)

    if not validate_sql(sql):
        return "Generated unsafe or invalid SQL."

    result = execute_sql(sql)

    summary_prompt = f"""
    You are a senior retail business insights analyst.

    Using the query result below, produce a professional analytical explanation
    that answers the user's question:

    QUESTION:
    {question}


    RESULT SUMMARY:
    number of rows: {len(result)}
    columns: {list(result.columns)}
    sample values: {result.head(5).to_dict(orient='records')}

    RESPONSE REQUIREMENTS:
    write EXACTLY 5 sentences (single paragraph)
    include numeric evidence from the result
    state what the numbers indicate
    highlight business meaning
    maintain confident analytical tone
    do NOT say "based on limited data"
    do NOT refuse to answer
    do NOT hallucinate external information
    """

    summary = model.generate_content(summary_prompt).text

    return {
        "sql": sql,
        "result": result,
        "summary": summary
    }
````

**`RAG Contextual Enrichment`**

````python
import pandas as pd

# Load Cloud Warehouse Comparison data
cloud_df = pd.read_csv("Cloud Warehouse Compersion Chart.csv")

# Load Expense IIGF data
expense_df = pd.read_csv("Expense IIGF.csv")

cloud_df.head(), expense_df.head()
````

````python
def df_to_chunks(df):
    chunks = []
    for _, row in df.iterrows():
        text = " | ".join([f"{col}: {row[col]}" for col in df.columns])
        chunks.append(text)
    return chunks
````

````python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer("all-MiniLM-L6-v2")

warehouse_chunks = df_to_chunks(cloud_df)
expense_chunks = df_to_chunks(expense_df)

warehouse_embeddings = embedder.encode(warehouse_chunks)
expense_embeddings = embedder.encode(expense_chunks)
````

````python
cloud_df.columns = cloud_df.columns.str.strip().str.lower().str.replace(" ", "_")
expense_df.columns = expense_df.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
def df_to_chunks(df):
    chunks = []
    for _, row in df.iterrows():
        text = " | ".join([f"{col}: {row[col]}" for col in df.columns])
        chunks.append(text)
    return chunks

warehouse_chunks = df_to_chunks(cloud_df)
expense_chunks = df_to_chunks(expense_df)
````

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

embedder = SentenceTransformer("all-MiniLM-L6-v2")

warehouse_embeddings = embedder.encode(warehouse_chunks)
expense_embeddings = embedder.encode(expense_chunks)
````

````python
def rag_search(query, embeddings, chunks):
    q_emb = embedder.encode([query])
    sims = cosine_similarity(q_emb, embeddings)[0]
    top_idx = sims.argmax()
    return chunks[top_idx]

def rag_answer(query, source_text):
    prompt = f"""
You are an insights analyst responding to a business question using ONLY the reference text provided.

### Allowed Behavior
Base the answer strictly on the reference text
Include numeric evidence, trends, comparisons, and rankings when present
Keep the tone professional, concise, and analytical
Summarize clearly in 3–5 sentences

### Forbidden Behavior
Do NOT invent data or assumptions
Do NOT answer from outside knowledge
Do NOT speculate if numbers are missing
Do NOT restate the question as the answer

### If the reference lacks enough information:
Respond with:
"Based on the provided data, there is not enough information to answer this question."

---

### Reference Data:
{source_text}

### Question:
{query}

### Produce:
A professional insight-based answer grounded ONLY in the reference.
"""
    return model.generate_content(prompt).text.strip()


def ask_rag(query, dataset="warehouse"):
    if dataset == "warehouse":
        src = rag_search(query, warehouse_embeddings, warehouse_chunks)
    else:
        src = rag_search(query, expense_embeddings, expense_chunks)

    answer = rag_answer(query, src)
    return answer
````

````python
ask_rag("Which warehouse has cheaper outbound fees?", "warehouse")
````

````python
ask_rag("Summarize the major expenses in the report", "expense")
````
